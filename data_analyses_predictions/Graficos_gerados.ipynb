```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Carregar o conjunto de dados
path = "/content/dataset_tipo_1.csv"
data = pd.read_csv(path)

data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Criar um DataFrame
df = data

# Gráfico de linha para o comprimento das folhas
plt.figure(figsize=(10, 8))
for id_alface in df['Id_alface'].unique():
    alface_data = df[df['Id_alface'] == id_alface]
    plt.plot(alface_data['Dia'], alface_data['Comprimento_folha'], label=f'Alface {id_alface}', marker = 'o', markersize = 6, linewidth = 2)

plt.title('Crescimento do Comprimento das Folhas ao Longo dos Dias')
plt.xlabel('Dia')
plt.ylabel('Comprimento das Folhas')
plt.legend(title='ID do Alface')
plt.grid(False)
plt.show()
```

```python
# Gráfico de linha para a altura das folhas
plt.figure(figsize=(10, 7))
for id_alface in df['Id_alface'].unique():
    alface_data = df[df['Id_alface'] == id_alface]
    plt.plot(alface_data['Dia'], alface_data['Altura_folha'], label=f'Alface {id_alface}', marker = 'o', markersize = 6, linewidth = 2)

plt.title('Crescimento da Altura das Folhas ao Longo dos Dias')
plt.xlabel('Dia')
plt.ylabel('Altura das Folhas')
plt.legend(title='ID do Alface')
plt.grid(False)
plt.show()
```

```python
# Gráfico de linha para a quantidade de folhas
plt.figure(figsize=(10, 6))
for id_alface in df['Id_alface'].unique():
    alface_data = df[df['Id_alface'] == id_alface]
    plt.plot(alface_data['Dia'], alface_data['Quant_folhas'], label=f'Alface {id_alface}', marker = 'o', markersize = 6, linewidth = 2)

plt.title('Crescimento de Folhas ao Longo dos Dias')
plt.xlabel('Dia')
plt.ylabel('Quantidade de Folhas')
plt.legend(title='Id do Alface')
plt.grid(False)
plt.show()
```

```python
import seaborn as sns

# Gráfico de dispersão para Quantidade de Folhas vs Comprimento das Folhas
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Quant_folhas', y='Comprimento_folha', hue='Id_alface', palette='Set1')
plt.title('Quantidade de Folhas vs Comprimento das Folhas')
plt.xlabel('Quantidade de Folhas')
plt.ylabel('Comprimento das Folhas')
plt.legend(title='Id do Alface')
plt.grid(True)
plt.show()
```

```python
# Gráfico de dispersão para Quantidade de Folhas vs Altura das Folhas
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Quant_folhas', y='Altura_folha', hue='Id_alface', palette='Set1')
plt.title('Quantidade de Folhas vs Altura das Folhas')
plt.xlabel('Quantidade de Folhas')
plt.ylabel('Altura das Folhas')
plt.legend(title='Id do Alface')
plt.grid(True)
plt.show()
```

```python
# Gráfico de barras para Quantidade de Folhas por ID de Alface
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Id_alface', y='Quant_folhas', ci=None)
plt.title('Quantidade de Folhas por ID de Alface')
plt.xlabel('ID do Alface')
plt.ylabel('Quantidade de Folhas')
plt.grid(False)
plt.show()
```

```python
# Gráfico de barras para Comprimento das Folhas por ID de Alface
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Id_alface', y='Comprimento_folha', ci=None)
plt.title('Comprimento das Folhas por ID de Alface')
plt.xlabel('ID do Alface')
plt.ylabel('Comprimento das Folhas')
plt.grid(False)
plt.show()
```

```python
# Gráfico de barras para Quantidade de Folhas por Dia, agrupado por ID de Alface
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Dia', y='Quant_folhas', hue='Id_alface', palette='Set1')
plt.title('Quantidade de Folhas por Dia de cada Alface')
plt.xlabel('Dia')
plt.ylabel('Quantidade de Folhas')
plt.legend(title='Id do Alface')
plt.grid(False)
plt.show()
```

```python
# Gráfico de barras para Quantidade de Folhas por Dia, agrupado por ID de Alface
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Dia', y='Comprimento_folha', hue='Id_alface', palette='Set1')
plt.title('Comprimento das Folhas por Dia para Cada ID de Alface')
plt.xlabel('Dia')
plt.ylabel('Comprimento da Folha')
plt.legend(title='ID do Alface')
plt.grid(False)
plt.show()
```

```python
# Gráfico de barras para Quantidade de Folhas por Dia, agrupado por ID de Alface
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Dia', y='Altura_folha', hue='Id_alface', palette='Set1')
plt.title('Altura das Folhas por Dia para Cada ID de Alface')
plt.xlabel('Dia')
plt.ylabel('Altura da Folha')
plt.legend(title='Id do Alface')
plt.grid(False)
plt.show()
```
